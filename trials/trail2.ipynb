```python
# importing essential libraries

import os
from requests import request
from six.moves import urllib
import pandas as pd
```

```python
# Download data From source

src_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/concrete/compressive/Concrete_Data.xls'

download_path = os.path.join("dataset","stength")

os.makedirs(download_path,exist_ok=True)

xls_file_path = os.path.join(download_path, "Concrete_Data.xls")

urllib.request.urlretrieve(src_url,xls_file_path)
```

```python
# Insert Data Into Table from CSV to Database TABLE

with open(file, mode='r') as f:
    next(f)
    reader = csv.reader(f, delimiter=',')
    for i in reader:
        data = ','.join([value for value in i])

        session.execute("USE {keyspace};".format(keyspace=keyspace))
        # session.execute('COPY {table_name}(Cement, Blast_Furnace_Slag , Fly_Ash , Water, Superplasticizer, Coarse_Aggregate, Fine_Aggregate, Age , Concrete_compressive_strength)  FROM  {file} | STDIN  VALUES ({data});').format(table_name=table_name,file=file, data=data)

        
        session.execute('INSERT INTO {table_name} (Cement, Blast_Furnace_Slag , Fly_Ash , Water, Superplasticizer, Coarse_Aggregate, Fine_Aggregate, Age , Concrete_compressive_strength) VALUES ({data});'.format(table_name=table_name, data=data))
```
